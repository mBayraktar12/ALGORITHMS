```python
def thief(n,stolenItems=0):
  if n == 0:
    return stolenItems
  elif n < 5 and n >= 3:
    stolenItems = stolenItems + 1
    return thief(n-3,stolenItems)
  elif n < 3:
    stolenItems = stolenItems + 1
    return thief(n-1,stolenItems)
  else:
    stolenItems = stolenItems + 1
    return thief(n-5,stolenItems)
```

```python
%time thief(100)
```
