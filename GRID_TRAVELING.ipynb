```python
def rec_gt(n,m):
  if (n == 1 or m == 1):
    return min(n,m)
  else:
    return rec_gt(n-1,m)+rec_gt(n,m-1)
```

```python
%time rec_gt(4,4)
```

```python
def memo_gt(n, m, memo={}):
    if (n, m) in memo:
        return memo[(n, m)]
    if (n == 1 or m == 1):
        return min(n,m)
    else:
        result = memo_gt(n-1, m, memo) + memo_gt(n, m-1, memo)
        memo[(n, m)] = result
        return result
```

```python
%time memo_gt(5,5)
```
